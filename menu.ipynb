```python
import pandas as pd
import numpy as np

# Copy to avoid altering original
df_num = music_df_cleaned.copy()

# 1. Extract numeric year from Date column
df_num['Year'] = pd.to_datetime(df_num['Date'], errors='coerce').dt.year

# 2. Text length features
df_num['Title_Length'] = df_num['Title'].astype(str).str.len()
df_num['Subtitle_Length'] = df_num['Subtitle'].astype(str).str.len()
df_num['Creator_Count'] = df_num['Creator'].astype(str).apply(lambda x: len(x.split(',')) if isinstance(x, str) else 0)
df_num['AdditionalCreators_Count'] = df_num['AdditionalCreators'].astype(str).apply(lambda x: len(x.split(',')) if isinstance(x, str) else 0)
df_num['Notes_Length'] = df_num['Notes'].astype(str).str.len()
df_num['PhysicalDescription_Length'] = df_num['PhysicalDescription'].astype(str).str.len()
df_num['Publisher_Length'] = df_num['Publisher'].astype(str).str.len()
df_num['Place_Length'] = df_num['Place'].astype(str).str.len()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute correlation on numeric columns
corr_matrix = df_num.corr(numeric_only=True)

plt.figure(figsize=(10, 8))

mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

sns.heatmap(
    corr_matrix,
    mask=mask,
    annot=True,
    cmap='coolwarm',
    vmin=-1,
    vmax=1,
    fmt=".2f"
)

plt.title('Music Metadata Correlation Heatmap', fontsize=16)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

selected_features = [
    'Year',
    'Title_Length',
    'Subtitle_Length',
    'Notes_Length',
    'Creator_Count',
    'AdditionalCreators_Count'
]

# Use only rows with valid Language values
df_lang = df_num.dropna(subset=['Language'])

sns.pairplot(
    df_lang[selected_features + ['Language']], 
    hue='Language', 
    diag_kind='kde',
    corner=True,
    palette='tab10',
    plot_kws={'alpha': 0.6, 's': 40}
)

plt.suptitle("Music Metadata Relationships by Language", y=1.02, fontsize=16)
plt.show()
```

### Clean & Prepare Both Datasets

```python
import pandas as pd

# Copy to avoid altering originals
music = music_df_cleaned.copy()
menu = menu_df_cleaned.copy()

# Add a label for dataset source
music["Collection"] = "Music"
menu["Collection"] = "Menu"

# Convert Date → Year
for df in [music, menu]:
    df["Year"] = pd.to_datetime(df["Date"], errors="coerce").dt.year

# Combine both datasets
combined = pd.concat([music, menu], ignore_index=True)

# Map language codes to readable names
lang_map = {
    "ger": "German",
    "zxx": "Instrumental / No Language",
    "fre": "French",
    "ita": "Italian",
    "eng": "English"
}
combined["Language_Name"] = combined["Language"].map(lang_map).fillna(combined["Language"])
```
