### Clean & Prepare Both Datasets

```python
import pandas as pd

# Copy to avoid altering originals
music = music_df_cleaned.copy()
menu = menu_df_cleaned.copy()

# Add a label for dataset source
music["Collection"] = "Music"
menu["Collection"] = "Menu"

# Convert Date → Year
for df in [music, menu]:
    df["Year"] = pd.to_datetime(df["Date"], errors="coerce").dt.year

# Combine both datasets
combined = pd.concat([music, menu], ignore_index=True)

# Map language codes to readable names
lang_map = {
    "ger": "German",
    "zxx": "Instrumental / No Language",
    "fre": "French",
    "ita": "Italian",
    "eng": "English"
}
combined["Language_Name"] = combined["Language"].map(lang_map).fillna(combined["Language"])
```

### Read another CSV with date till 1915

```python
music_df_1915 = pd.read_csv('metadata_pre_1915.csv')

music_df_1915.head(30)
```

```python
import pandas as pd
import re

# Assume your dataframe is called df (either menu_df_cleaned or music_df_cleaned)
df = music_df_1915.copy()

def expand_year_ranges(row):
    """
    For a given row, if Date contains a range like '1915-1918',
    it will return a list of years [1915, 1916, 1917, 1918].
    Otherwise, returns the single year.
    """
    text = str(row["Date"])
    years = re.findall(r"(1[0-9]{3}|20[0-9]{2})", text)

    if len(years) == 2:
        start, end = map(int, years)
        return list(range(start, end + 1))
    elif len(years) == 1:
        return [int(years[0])]
    else:
        return [None]  # If no valid year found

# Apply expansion logic
expanded_rows = []

for _, row in df.iterrows():
    for year in expand_year_ranges(row):
        new_row = row.copy()
        new_row["Year"] = year
        expanded_rows.append(new_row)

# Create new expanded DataFrame
expanded_df = pd.DataFrame(expanded_rows)

# Optional: drop rows without valid year
expanded_df = expanded_df.dropna(subset=["Year"]).reset_index(drop=True)

#print(expanded_df[["Date", "Year"]].head(15))
expanded_df.head()
expanded_df.info()
```

### Renaming the values

```python
import pandas as pd

# List of columns to clean (update with your actual text columns)
columns_to_clean = ['Place','Publisher','AdditionalCorporateCreators','PPN','Title','Subtitle','Creator','AdditionalCreators','Language','Notes','PhysicalDescription','Date','Shelfmark_Original','PPN_Original','Year']

def clean_text_column(series):
    """
    Cleans a text column by:
    - Removing square brackets []
    - Removing quotes '
    - Stripping extra whitespace
    """
    return (
        series.astype(str)        # Ensure string type
              .str.strip()        # Remove leading/trailing spaces
              .str.strip('[]')    # Remove square brackets
              .str.strip('"\'')   # Remove quotes
    )

# Apply cleaning to all listed columns
for col in columns_to_clean:
    if col in expanded_df.columns:
        expanded_df[col] = clean_text_column(expanded_df[col])

# Optional: Standardize capitalization
expanded_df[columns_to_clean] = expanded_df[columns_to_clean].apply(lambda x: x.str.title())
expanded_df.head(50)
```
